# Encryption
 2 simple steps to set up encryption
 1. generate a key

```python
key = Fernet.generate_key()
print('key: ', key) #key:  b'fGodsOVEePr-Lw0_weTHTW0621psgS6e0kjR9babP7M=' (it keeps changing every time I run the code)
```

2. create an object using the key

```python
cipher = Fernet(key)
print('cipher: ', cipher) # cipher:  <cryptography.fernet.Fernet object at 0x000001CCA8534050> (it also keeps changing)
```

```python
# print(dir(cipher)) #`ve alot
```

use .encrypt() to implement encryption

```python
encrypted = cipher.encrypt('hi'.encode()) # data must be bytes thats why added .encode()
print("Encrypted:", encrypted) #b'gAAAAABoM1fTYTm6h4mbs7UxWSoVHjXN2M7Zb57qkhHt9DIykV_UOyjT0OZCiCQdqV62KmiAVD0vLPA2UTez6FbIXFw3lvJOSw==' (it also)
```

# Now to decrypt

```python
decrypted = cipher.decrypt(encrypted).decode() # used .decode() to turn it back into string from byte
print("Decrypted:", decrypted)
```

**Hashing vs Encryption**:

| Feature         | Hashing                                 | Encryption                         |
| --------------- | --------------------------------------- | ---------------------------------- |
| üîÅ Reversible?  | ‚ùå No (one-way)                          | ‚úÖ Yes (two-way: encrypt/decrypt)   |
| üîê Purpose      | Verify data integrity (e.g., passwords) | Protect data confidentiality       |
| üß† Output       | Fixed-length hash (e.g., SHA-256)       | Variable cipher text (Fernet, AES) |
| üóùÔ∏è Key Needed? | ‚ùå No                                    | ‚úÖ Yes (for both encrypt/decrypt)   |

**Hash = fingerprint**

**Encryption = secret message with a key**.


## When should I hash?
Use **hashing** when you only need to **verify** something and **never need to recover** the original data.  
‚úÖ Example: Passwords, ATM PINs, verification tokens.

## When should I encrypt?
Use **encryption** when you need to **recover or read** the original data later.  
‚úÖ Example: Messages, stored notes, sensitive files.


* WhatsApp chat ‚Üí needs to be readable ‚Üí encrypt
* ATM PIN ‚Üí just verify match ‚Üí hash
